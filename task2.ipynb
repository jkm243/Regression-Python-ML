```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Load the data into a pandas dataframe
data = pd.read_csv('final_test_regression_data.csv')
```

```python
# Split the data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
# Create two linear regression models
regressor1 = LinearRegression()
regressor2 = LinearRegression()
```

```python
# Train the first model on the training data
X_train = train_data[['x']]
y_train = train_data['y']
regressor1.fit(X_train, y_train)
```

```python
# Train the second model on a different feature of the training data
X_train2 = train_data[['x']]
regressor2.fit(X_train2, y_train)
```

```python
# Predict the target variable on the testing data using the trained models
X_test = test_data[['x']]
y_test = test_data['y']
y_pred1 = regressor1.predict(X_test)
y_pred2 = regressor2.predict(test_data[['x']])
```

```python
# Plot the predictions and the actual values of the testing data
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred1, color='blue', label='Regressor 1')
plt.plot(test_data[['x']], y_pred2, color='red', label='Regressor 2')
plt.legend()
plt.show()
```

```python
# Plot the points of the training data
plt.scatter(train_data[['x']], train_data[['y']], color='black')
plt.show()
```

```python
# Plot the predictions of the first regressor inside and outside the range of values of the training data
x_range = pd.DataFrame({'x': range(int(train_data[['x']].min()), int(train_data[['x']].max())+1)})
y_pred_range = regressor1.predict(x_range)
plt.scatter(train_data[['x']], train_data[['y']], color='black')
plt.plot(x_range, y_pred_range, color='blue', label='Regressor 1')
plt.legend()
plt.show()
```
